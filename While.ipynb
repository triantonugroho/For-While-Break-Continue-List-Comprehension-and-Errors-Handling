While

A while loop statement in Python programming language repeatedly executes a target statement as long as a given condition is true.

Conditions can be any expression, and please keep in mind that True in Python includes all non-zero values.

When the condition is False, the Program will proceed to the statement block line.

Example

```python
a = int(input())

while(a < 3):
  print("number {}".format(a))
  a = a + 0.5 
```

```python
b = int(input())

while(b < 10):
  b = b + 2
  print("number {}".format(b))
```

While loops with Else

The else part is executed if the condition in the while loop evaluates to False.

The while loop can be terminated with a break statement. In such cases, the else part is ignored. Hence, a while loop's else part runs if no break occurs and the condition is false.

Example

```python
number = int(input())

while (number>0):
  number = number - 2
  print(number)
else:
  print("finished")
```

```python
number = int(input())

while (number < 15):
  number = number + 2
  if number == 17:
    break
  print(number)
else:
    print("finished")
```

```python
i = 2
while i < 7:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```python
# Program to add natural
# numbers up to 
# sum = 1+2+3+...+n

# To take input from the user,
# n = int(input("Enter n: "))

n = 7

# initialize sum and counter
sum = 0
i = 1

while i <= n:
    sum = sum + i
    i = i+1    # update counter

# print the sum
print("The sum is", sum)
```

In the above program, the test expression will return True as long as our counter variable i is less than or equal to n (7 in our program).

We need to increase the value of the counter variable in the loop body. This is very important (and mostly forgotten). Failing to do so will result in an infinite loop (an endless loop).
